```python
import numpy as np
import pandas as pd
from si_prefix import si_format
```

```python
# Path of the file to read
files = ['CLK', 'A3', 'A2', 'A1', 'A0', 'B3', 'B2', 'B1', 'B0', 'P7', 'P6', 'P5', 'P4', 'P3', 'P2', 'P1', 'P0']

df = pd.DataFrame()
for file in files:
    add = pd.read_csv('./CSV/' + file + '.csv')
    df = pd.concat([df, add], axis=1)
```

```python
df = df.drop(df.columns[[i for i in range(len(df.columns)) if i % 2 == 0 and i != 0]], axis=1)
df
```

```python
quantized = np.digitize(df.iloc[:,1:], bins=[0.425])
q_df = pd.DataFrame(quantized, columns=df.columns[1:])
```

```python
q_df = pd.concat([df.iloc[:,0:1], q_df], axis=1)
q_df.rename(columns = {'clk (interpolated wth step=1e-10) X':'time '}, inplace = True)
q_df.rename(columns = lambda x: x[0:x.find(' (')], inplace = True)
q_df
```

```python
q_df['A_bin'] = q_df['a3'].astype(str) + q_df['a2'].astype(str) + q_df['a1'].astype(str) + q_df['a0'].astype(str)
q_df['B_bin'] = q_df['b3'].astype(str) + q_df['b2'].astype(str) + q_df['b1'].astype(str) + q_df['b0'].astype(str)
q_df['P_bin'] = q_df['p7'].astype(str) + q_df['p6'].astype(str) + q_df['p5'].astype(str) + q_df['p4'].astype(str) + q_df['p3'].astype(str) + q_df['p2'].astype(str) + q_df['p1'].astype(str) + q_df['p0'].astype(str)
q_df['A'] = q_df['A_bin'].apply(lambda x: int(x, 2))
q_df['B'] = q_df['B_bin'].apply(lambda x: int(x, 2))
q_df['P'] = q_df['P_bin'].apply(lambda x: int(x, 2))
q_df
```

```python
input_arr = []
curr_clk = 0

for row in q_df.iterrows():
    if (curr_clk == 0 and row[1]['clk'] == 1):
        input_arr.append([row[1]['A'], row[1]['B'], row[1]['time']])
        if (row[1]['A'] == 15 and row[1]['B'] == 15):
            print(row[1]['A'])
            break
    curr_clk = row[1]['clk']
```

```python
delay_df = pd.DataFrame(input_arr, columns=['A', 'B', 'input_time'])
delay_df.insert(3, 'output_time', 0)
delay_df.insert(4, 'delay', 0)
delay_df
```

```python
curr_A = 0
curr_B = 0

for row in q_df.iterrows():
    if (row[1]['clk'] == 1):
        if (row[1]['P'] == curr_A * curr_B and (row[1]['A'] != curr_A or row[1]['B'] != curr_B)):
            delay_df.loc[(delay_df['A'] == curr_A) & (delay_df['B'] == curr_B), 'output_time'] = row[1]['time']
            delay_df.loc[(delay_df['A'] == curr_A) & (delay_df['B'] == curr_B), 'delay'] = \
                row[1]['time'] - delay_df.loc[(delay_df['A'] == curr_A) & (delay_df['B'] == curr_B), 'input_time']
            curr_B += 1
        if (curr_B == 16):
            curr_A += 1
            curr_B = 0
delay_df
```

```python
delay_df['delay_si_format'] = delay_df['delay'].apply(lambda x: si_format(x))
delay_df
```

```python
delay_df.to_csv('./delay.csv', index=False)
```
